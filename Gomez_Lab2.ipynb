10. Generator Does not Run Immediately

```python
def demo():
  print("A")
  yield 1
  print("B")
x = demo()
```

11. Generator Resumes

```python
g = demo()
print(next(g))
print(next(g))
```

12. Closures Preserve Values

```python
def outer():
  x = 10
  def inner():
    return
  return inner
f = outer()
print(f())
```

13. Decorators Change Behaviour

```python
def deco(fn):
  def wrapper():
    print("before")
    fn()
    print("after")
  return wrapper
def hello():
  print("hello")
hello()
```

14. Attribute Look up

```python
class A:
  x = 10
a = A()
print(a. x)
```

15. Duck Typing

```python
class MyRange:
  def __iter__(self):
    return self
  def __next__(self):
    raise StopIteration
  for x in MyRange():
    print(x)
```
