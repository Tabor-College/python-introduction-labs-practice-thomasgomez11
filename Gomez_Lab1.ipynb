Example 7: Rebinding vs Mutation

```python
def reset_wrong(lst):
  lst= []

def reset_right(lst):
  lst.clear()

  data = [1, 2, 3]
  reset_wrong(data)
  print(data)

  reset_right(data)
  print(data)
```

Reflection Questions:

Was the object mutable or immutable?
Mutable


Did Python mutate the object or rebind the name? rebound the name


Did the change affected other references?
Yes
